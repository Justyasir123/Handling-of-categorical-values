```python
import pandas as pd
```

```python
df=pd.read_csv('customer.csv')
df
```

**ORDINAL ENCODING-FOR INDEPENDANT VARIABLE**

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
review_order=['Poor','Average','Good']
education_order=['School','UG','PG']
encoder=OrdinalEncoder(categories=[review_order,education_order])
```

```python
df[['review','education']]=encoder.fit_transform(df[['review','education']])
```

```python
df.head()
```

**LABEL ENCODING- FOR TARGET VARIABLE**

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoder=LabelEncoder()
```

```python
df['purchased']=encoder.fit_transform(df['purchased'])
```

```python
df.head()
```
